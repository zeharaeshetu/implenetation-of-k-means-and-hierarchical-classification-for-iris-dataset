# load important libraries

```python
from sklearn import datasets
import sklearn
from sklearn.decomposition import PCA
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering 
import seaborn as sns
```

# load and normalize the dataset 

```python
from sklearn.preprocessing import scale
iris= datasets.load_iris()
data = scale(iris.data)
target= pd.DataFrame(iris.target)
variable_names= iris.feature_names
data[0:50]
```

# Visualizing the data before clustering

```python
plt.scatter(data[:,0],data[:,1],c='black')
plt.show()
```

# Hierarchical Clustering


# Find Opimum k 

```python


# Importing Hierarchical Clustering data from scipy
from scipy.cluster.hierarchy import dendrogram,linkage
Z= linkage(data,'ward')
max_d= 9.08


plt.figure(figsize=(20,7))
plt.title('Dendrogram for Iris Dataset Clustering')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
dendrogram(
Z,truncate_mode='lastp', p=50,leaf_rotation=90.,leaf_font_size=8.,)
plt.axhline(y=max_d, c='k')
plt.show()
```

# generate the hierarchical cluster 

```python
from sklearn.cluster import AgglomerativeClustering 
k=3
HClustering=AgglomerativeClustering(n_clusters = k, affinity = 'euclidean', linkage = 'complete') 
HClustering.fit_predict(data)

```

```python
from sklearn.cluster import AgglomerativeClustering
k=3
hc= AgglomerativeClustering(k,affinity='euclidean',linkage='complete')
y_hc=hc.fit(data)
```
